```python
import networkx as nx
import matplotlib.pyplot as plt
from graphs_for_slides import *
```

```python
g = nx.Graph()
n=iter(range(10))
g.add_edges_from(zip(n,n))
layout = K_2_xn_layout(5)
plt.axes().set_aspect('equal', adjustable='datalim')
nx.draw(g, pos=layout, with_labels=True, font_weight='bold', font_color='white')
```

```python
for el in nx.connected_components(g): print(el)
```

```python
g1=g.subgraph(el)
ax = plt.subplot(121)
nx.draw(g, pos=layout, with_labels=True, font_weight='bold', font_color='white')
ax = plt.subplot(122)
nx.draw(g1, pos=layout, with_labels=True, font_weight='bold', font_color='white', node_color='magenta')
```

```python
g = nx.Graph()
nx.add_star(g, range(4))
nx.add_cycle(g, range(4,10))
nx.add_path(g, range(10,13))
nx.draw_shell(g, with_labels=True, font_weight='bold', font_color='white')
```

```python
g = nx.Graph()
nx.add_star(g, range(4))
nx.add_cycle(g, range(4,10))
nx.add_path(g, range(5,13))
nx.draw_shell(g, with_labels=True, font_weight='bold', font_color='white')
```

```python
g1 = nx.complete_graph(3)
g2 = nx.cycle_graph(4)
ax = plt.subplot(121)
ax.set_xmargin(0.1)
ax.set_aspect('equal', adjustable='datalim')
nx.draw_shell(g1, with_labels=True, font_weight='bold', font_color='white')
ax = plt.subplot(122)
ax.set_xmargin(0.1)
ax.set_aspect('equal', adjustable='datalim')
nx.draw_shell(g2, with_labels=True, font_weight='bold', font_color='white', node_color='magenta')
```

```python
g = nx.compose(g1,g2)
plt.axes().set_aspect('equal', adjustable='datalim')
nx.draw_shell(g, with_labels=True, font_weight='bold', font_color='white')
```

```python
g = nx.disjoint_union(g1,g2)
plt.axes().set_aspect('equal', adjustable='datalim')
nx.draw_shell(g, with_labels=True, font_weight='bold', font_color='white')
```

```python
C4_complement = nx.complement(nx.cycle_graph(4))
plt.axes().set_aspect('equal', adjustable='datalim')
nx.draw_shell(C4_complement, with_labels=True, font_weight='bold', font_color='white')
```

```python
g1 = nx.complete_bipartite_graph(3,3)
g2 = g1.copy()
g2.remove_edge(1,4)
pos = nx.bipartite_layout(g1,(v for v, prop in g1.nodes('bipartite') if prop==0), align='horizontal')
ax=plt.subplot(121); ax.set_aspect('equal', adjustable='datalim'); ax.set_xmargin(0.1)
nx.draw(g1, pos=pos, with_labels=True, font_weight='bold', font_color='white')
ax=plt.subplot(122); ax.set_aspect('equal', adjustable='datalim'); ax.set_xmargin(0.1)
nx.draw(g2, pos=pos, with_labels=True, font_weight='bold', font_color='white', node_color='magenta')
```
